```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
# import matplotlib.dates as mdates

import pandas as pd
from datetime import datetime

# font 설정 D2coding
d2_path = "./ref/D2Coding-Ver1.3.2-20180524.ttf"
fm.fontManager.addfont(d2_path)
matplotlib.rcParams["font.family"] = "D2coding"

file_path = "./ref/온라인쇼핑몰_판매매체별_상품군별거래액.csv"
ori = pd.read_csv(file_path, encoding="EUC-KR") #utf-8
ori.columns = ["category","date","internet","mobile"]

data = ori
data['date'] = ori['date'].map(lambda x: x.rstrip(' p)')) #date에 있는 P)제거

#날짜 형식으로 바꾸기
ori['date'] = pd.to_datetime(data['date'], format = '%Y.%m')
ori['total'] = ori['internet'] + ori['mobile']

category_data = ori.copy()
obj = ['문화 및 레저서비스']
category_data = category_data[category_data['category'].isin(obj) ]

fig, ax = plt.subplots(figsize =(15,10))
date = category_data['date']
mobile = category_data['mobile']
internet = category_data['internet']
total = category_data['total']

ax.plot(date, total, label='총합', linestyle='-', marker='*', markersize=10)
ax.scatter(date, mobile, s=20, label='문화 및 레저 서비스(모바일)')
ax.scatter(date, internet, s=20, label='문화 및 레저 서비스(인터넷)')
ax.legend(loc="best")
plt.show()
print('''
    문화 및 레저 서비스는 계절에 따른 변동이 있으나 년도별로 오르는 추세를 보였지만, 2020년 코로나 사태로 인해 급격한 매출 감소를 보인다. 2022년
    하반기부터 다시 증가하는 추세를 보이며 계절에 따른 변동또한 년도별로 비슷한 지표를 보인다.
''')
```
